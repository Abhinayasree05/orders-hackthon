```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("restaurants.db")
print("Database connected")
```

```python
try:
    with open("restaurants.sql", "r", encoding="utf-8") as f:
        sql_script = f.read()

    conn.executescript(sql_script)
    print("restaurants.sql executed successfully")

except Exception as e:
    print("SQL already executed or skipped safely")
    print(e)
```

```python
restaurants = pd.read_sql(
    "SELECT * FROM restaurants",
    conn
)

restaurants.head()
```

```python
restaurants.columns
```

```python
orders_users = orders.merge(
    users,
    on="user_id",
    how="left"
)
```

```python
final_df = orders_users.merge(
    restaurants,
    on="restaurant_id",
    how="left"
)
```

```python
final_df.head()
```

```python
final_df.shape
```

```python
final_df.to_csv("final_food_delivery_dataset.csv", index=False)
```

```python
final_df["membership"] = final_df["membership"].fillna("Unknown")
final_df["city"] = final_df["city"].fillna("Unknown")
final_df["cuisine"] = final_df["cuisine"].fillna("Unknown")
final_df.isnull().sum()
```

```python
final_df["order_date"] = pd.to_datetime(
    final_df["order_date"],
    format="%d-%m-%Y"
)
```

```python
final_df["year"] = final_df["order_date"].dt.year
final_df["month"] = final_df["order_date"].dt.month
final_df["quarter"] = final_df["order_date"].dt.to_period("Q").astype(str)
```

```python
final_df.to_csv("final_food_delivery_dataset.csv", index=False)
```
